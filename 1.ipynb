```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = ''.join(c.lower() for c in s if c.isalnum())
        left = 0 
        right = len(s) - 1

        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1

        return True
```

```python
from typing import List

class Solution:
    def maxDistToClosest(self, seats: List[int]) -> int:
        max_zero_counter = -10**11
        zs = zf = 0
        for i in range(len(seats)):
            if seats[i] == 1:
                break
            else:
                zs += 1
        for i in range(len(seats)-1, 0, -1):
            if seats[i] == 1:
                break
            else:
                zf += 1
        zero_counter = 0
        for i in range(len(seats)):
            if seats[i] == 0:
                zero_counter += 1
                if zero_counter > max_zero_counter:
                    max_zero_counter = zero_counter
            else:
                zero_counter = 0
        if max_zero_counter % 2 == 1:
            return max(max_zero_counter // 2 + 1, zs, zf)
        else:
            return max(max_zero_counter // 2 , zs, zf)

solution = Solution()
test_seats = [1,0,0,0,1,0,1]
result = solution.maxDistToClosest(test_seats)
print(f"Результат для {test_seats}: {result}")
```
